```python
sns.heatmap(df[['Attrition','DistanceFromHome','YearsSinceLastPromotion','YearsWithCurrManager', 'YearsAtCompany', 'OverTime','JobSatisfaction', 'PercentSalaryHike', 'Age']].corr(), cmap="rocket_r", annot=True, fmt='.2f')
```

```python
df_jobs = df[['JobRole','OverTime','PercentSalaryHike']].groupby(['JobRole', 'OverTime'], as_index=False).mean()
df_jobs
```

```python
df_jobs_pivot = df_jobs.pivot(index='JobRole', columns='OverTime')
df_jobs_pivot
```

```python
sns.heatmap(df_jobs_pivot, cmap="Blues")
```

#### Conclusão
Hora Extra aparenta não influenciar em um aumento de salaário para a maioria dos cargos

```python
#bins = np.linspace(min(df['Age']), max(df['Age']), 4)
bins = [18,27,43,59,60]
print(bins)
GroupNames = ['Generation Z','Generation Y','Generation X','Baby Boomers']
df['Generation'] = pd.cut(df['Age'], bins, labels=GroupNames, include_lowest=True)
```

```python
df.sample(5)
```

```python
px.box(data_frame=df, x='Generation', y='JobSatisfaction', color="EducationField", category_orders={"Generation": ['Baby Boomers','Generation X','Generation Y','Generation Z']})
```

```python
px.box(data_frame=df, x='Generation', y='EnvironmentSatisfaction', color="EducationField", category_orders={"Generation": ['Baby Boomers','Generation X','Generation Y','Generation Z']})
```

```python
df2 = df[['Generation','EducationField','MonthlyIncome']].groupby(['Generation', 'EducationField'], as_index=False, observed=False).mean()
df2
```

```python
df2_pivot = df2.pivot(columns="Generation", index="EducationField")
df2_pivot
```

```python
fig, ax = plt.subplots(figsize=(7,7), dpi=150)
sns.heatmap(df2_pivot, cmap="Greens", xticklabels=['GenZ','GenY','GenX','Baby Boomers'], annot=True, fmt='.0f')
```
